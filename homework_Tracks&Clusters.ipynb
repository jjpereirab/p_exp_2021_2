## Exercise 4 ##

What is "Particle PDG ID"?

**R/** The particle PDG ID is a unique code assigned by the [Particle Data Group](https://en.wikipedia.org/wiki/Particle_Data_Group) to each type of particle. Particles and antiparticles have assigned positive and negative numbers respectively. I have confirmed for some of the numbers that they are [here](https://pdg.lbl.gov/2007/reviews/montecarlorpp.pdf).

Without going into detail, I find interesting that the most general number assigned could have up to 7 digits and is composed by the following particle parameters

$$\pm n n_{r} n_{L} n_{q_{1}} n_{q_{2}} n_{q_{3}} n_{J}$$

```python
pdg_id_list = []
for event in tree:
    pdg_id_list += list(tree.Particles_pdgID)

print("min id: {} | max id: {}".format(min(pdg_id_list),max(pdg_id_list)))
```

The max 3334 and min -3334 ids correspond to strange baryons $\Omega^-$. Plotting a histogram only for the first event

```python
hPDGid = 0
hPDGid = ROOT.TH1F("hist_pPDGid","Particle PDG id for the first event; PDG id ; Amount", 200, -3344, 3344)

for i in range(tree.Particles_pdgID.size()):
    hPDGid.Fill(tree.Particles_pdgID.at(i))
        
hPDGid.SetFillColor(2)
hPDGid.Draw()
canvas.Draw()
```
